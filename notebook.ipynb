<font color= #004270> <font size = 5> 1.1 Variables con regresión simbólica </font>

<div style="text-align: justify">
    <font color= #004270> "La regresión simbólica (una aplicación de la Programación Genética) tiene el mismo objetivo de la regresión lineal pero con un espectro mucho mayor de búsqueda y muchas menos limitaciones. Dados los datos, buscará la expresión algebraica que identifique el comportamiento de estos accediendo a todo tipo de funciones y combinaciones algebraicas" (Moreno, 2007) </font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270> Para comprender el proceso de una regresión simbólica primero hay que conocer una regresión lineal multiple, que presentan las siguiente forma al pronosticar: </font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270>$$Yhat = a_1+a_2X_1+a_3X_2 ...a_nX_{n-1}$$ </font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270>De esta forma, usando un modelo de regresión lineal se puede llegar a un pronóstico con un bajo error, si hay una correcta ingeniería de variables. Sin embargo, es necesario saber que una regresión de este tipo tiene una gran limitación: no puede detectar combinaciones no lineales de las variables.
</font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270>Por el otro lado, uan regresión simbólica consiste en la itnegración de las mismas variables con un set de funciones dados, como por ejemplo senos, cosenos, valores absolutos, valores máximos y mínimos, divisiones, etc. Esta regresión genera una función por medio de un algoritmo genético, que use una combinación de las variables y las funciones anteriores. Un ejemplo de una función de una regresión simbólica podría ser el siguiente:
</font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270>$$Yhat_{simbolico} = \cos(X_n-a_1)-sen(\frac{X_1}{max(X_2,X_3)})$$ </font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270> Al ser un algoritmo genético, tiene desventajas que tienen que ver con máximos y mínimos locales(Moreno, 2007) y no hay garantía de que la ecuación encontrada sea la mejor, ya que se ven involucrados múmeros aleatorios
</font>
</div>
<br>
<div style="text-align: justify">
    <font color= #004270> Es por estas razones que en este proyecto se propone la unción de estas dos regresiones. Se utilizará la regresión simbólica para generar variables no lineales, y se añadirán al modelo de una regresión simbólica, para comparar la suma de los residuales al cuadrado, y bsucar el mejor modelo posible en el alcance del proyecto
</font>
</div>
